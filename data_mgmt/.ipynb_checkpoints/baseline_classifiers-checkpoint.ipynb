```python
import rasterio
from rasterio.mask import mask
import geopandas as gpd
import numpy as np
from shapely.geometry import mapping
```

```python
data_dir = '../data/'

# Bird Shoal
img_fp = data_dir + "mosaics/180522BirdSEO_WGS84_transparent_mosaic_group1.tif"
```

```python
full_dataset = rasterio.open(img_fp)
img_rows, img_cols = full_dataset.shape
img_bands = full_dataset.count
print(full_dataset.shape) # dimensions
print(full_dataset.count) # bands
```

```python
from rasterio.plot import show
from rasterio.plot import show_hist
from rasterio.windows import Window

import matplotlib.pyplot as plt

fig, axrgb = plt.subplots(1, 1, figsize=(10,10))
windowed_raster = full_dataset.read((1,2,3), window=Window(79000, 29000, 10000, 10000))
show(windowed_raster, transform=full_dataset.transform, ax=axrgb)
plt.show()
```
